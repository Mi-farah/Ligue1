```python
import pandas as pd
import os

os.getcwd()
```

```python
df_emissions = pd.read_csv("./backend/data/calculated_travels/total_emissions.csv")
```

```python
df_emissions.head()
```

```python
visiting_team = "Monaco"
host_team = "Paris Saint-Germain"
df_emissions[
    (df_emissions["visiting_team"] == visiting_team)
    & (df_emissions["hosting_team"] == host_team)
]
```

```python
df_emissions.columns
```

```python
scenarios_list_travels = []


for idx, row in df_emissions.iterrows():
    visiting_team = row["visiting_team"]
    hosting_team = row["hosting_team"]
    if visiting_team == hosting_team:
        continue
    original_transport_type = row["transport_type"]
    original_emissions_kg_co2 = row["emissions_kg_co2"]
    original_emissions_kg_co2_wo_empty_bus = row["emissions_kg_co2_wo_empty_bus"]
    original_travel_time_seconds = row["travel_time_seconds"]
    alternative_emissions_kg_co2 = row["alternative_emissions_kg_co2"]
    alternative_emissions_kg_co2_wo_empty_bus = row[
        "alternative_emissions_kg_co2_wo_empty_bus"
    ]
    alternative_travel_time_seconds = row["alternative_travel_time_seconds"]
    alternative_transport_type = row["alternative_transport_type"]
    # SA
    scenarios_list_travels.append(
        {
            "scenario": "SA",
            "visiting_team": visiting_team,
            "hosting_team": hosting_team,
            "transport_type": original_transport_type,
            "emissions_kg_co2": original_emissions_kg_co2,
            "travel_time_seconds": original_travel_time_seconds,
        }
    )
    scenarios_list_travels.append(
        {
            "scenario": "SA'",
            "visiting_team": visiting_team,
            "hosting_team": hosting_team,
            "transport_type": original_transport_type,
            "emissions_kg_co2": original_emissions_kg_co2_wo_empty_bus,
            "travel_time_seconds": original_travel_time_seconds,
        }
    )
    # if avion, we chang
    if original_transport_type == "avion":
        # S-3h
        if original_travel_time_seconds < 10800:
            emissions_kg_co2 = alternative_emissions_kg_co2
            travel_time_seconds = alternative_travel_time_seconds
            transport_type = alternative_transport_type
            emissions_kg_co2_wo_empty_bus = alternative_emissions_kg_co2_wo_empty_bus
        else:
            emissions_kg_co2 = original_emissions_kg_co2
            travel_time_seconds = original_travel_time_seconds
            transport_type = original_transport_type
            emissions_kg_co2_wo_empty_bus = original_emissions_kg_co2_wo_empty_bus
        scenarios_list_travels.append(
            {
                "scenario": "S-3h",
                "visiting_team": visiting_team,
                "hosting_team": hosting_team,
                "transport_type": transport_type,
                "emissions_kg_co2": emissions_kg_co2,
                "travel_time_seconds": travel_time_seconds,
            }
        )
        scenarios_list_travels.append(
            {
                "scenario": "S-3h'",
                "visiting_team": visiting_team,
                "hosting_team": hosting_team,
                "transport_type": transport_type,
                "emissions_kg_co2": emissions_kg_co2_wo_empty_bus,
                "travel_time_seconds": travel_time_seconds,
            }
        )

        # S-6h
        if original_travel_time_seconds < 21600:
            emissions_kg_co2 = alternative_emissions_kg_co2
            travel_time_seconds = alternative_travel_time_seconds
            transport_type = alternative_transport_type
            emissions_kg_co2_wo_empty_bus = alternative_emissions_kg_co2_wo_empty_bus
        else:
            emissions_kg_co2 = original_emissions_kg_co2
            travel_time_seconds = original_travel_time_seconds
            transport_type = original_transport_type
            emissions_kg_co2_wo_empty_bus = original_emissions_kg_co2_wo_empty_bus
        scenarios_list_travels.append(
            {
                "scenario": "S-6h",
                "visiting_team": visiting_team,
                "hosting_team": hosting_team,
                "transport_type": transport_type,
                "emissions_kg_co2": emissions_kg_co2,
                "travel_time_seconds": travel_time_seconds,
            }
        )
        scenarios_list_travels.append(
            {
                "scenario": "S-6h'",
                "visiting_team": visiting_team,
                "hosting_team": hosting_team,
                "transport_type": transport_type,
                "emissions_kg_co2": emissions_kg_co2_wo_empty_bus,
                "travel_time_seconds": travel_time_seconds,
            }
        )

        # S-NP
        if original_transport_type == "avion":
            emissions_kg_co2 = alternative_emissions_kg_co2
            travel_time_seconds = alternative_travel_time_seconds
            transport_type = alternative_transport_type
            emissions_kg_co2_wo_empty_bus = alternative_emissions_kg_co2_wo_empty_bus
        else:
            emissions_kg_co2 = original_emissions_kg_co2
            travel_time_seconds = original_travel_time_seconds
            transport_type = original_transport_type
            emissions_kg_co2_wo_empty_bus = original_emissions_kg_co2_wo_empty_bus
        scenarios_list_travels.append(
            {
                "scenario": "S-NP",
                "visiting_team": visiting_team,
                "hosting_team": hosting_team,
                "transport_type": transport_type,
                "emissions_kg_co2": emissions_kg_co2,
                "travel_time_seconds": travel_time_seconds,
            }
        )
        scenarios_list_travels.append(
            {
                "scenario": "S-NP'",
                "visiting_team": visiting_team,
                "hosting_team": hosting_team,
                "transport_type": transport_type,
                "emissions_kg_co2": emissions_kg_co2_wo_empty_bus,
                "travel_time_seconds": travel_time_seconds,
            }
        )
```

```python
df_scenarios = pd.DataFrame(scenarios_list_travels)
```

```python
# emissions per scenario per team
all_teams = set(df_scenarios["visiting_team"].unique().tolist()).union(
    df_scenarios["hosting_team"].unique().tolist()
)
print(all_teams)
```

```python
df_emissions_per_scenario_per_team_per_transport_type = pd.DataFrame()
for team in all_teams:
    df_team = df_scenarios[
        (df_scenarios["visiting_team"] == team) | (df_scenarios["hosting_team"] == team)
    ]
    df_team = (
        df_team.groupby(["scenario", "transport_type"])[
            ["emissions_kg_co2", "travel_time_seconds"]
        ]
        .sum()
        .reset_index()
    )
    df_team.insert(0, "team", team)
    df_emissions_per_scenario_per_team_per_transport_type = pd.concat(
        [df_emissions_per_scenario_per_team_per_transport_type, df_team]
    )
```

```python
df_emissions_per_scenario_per_team_per_transport_type.to_csv(
    "./backend/data/calculated_travels/emissions_per_scenario_per_team_per_transport_type.csv",
    index=False,
)
```

```python
df_emissions_per_scenario_per_team = (
    df_emissions_per_scenario_per_team_per_transport_type.groupby(["scenario", "team"])[
        ["emissions_kg_co2", "travel_time_seconds"]
    ]
    .sum()
    .reset_index()
)
```
