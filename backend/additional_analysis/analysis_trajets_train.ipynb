```python
from backend.global_variables import DATA_PATH
import pandas as pd
import ast
```

```python
train_trajets_df = pd.read_csv(DATA_PATH + "train_emissions.csv")
```

```python
train_trajets_df["route_details_dict"] = train_trajets_df["route_details"].apply(
    ast.literal_eval
)
```

```python
def check_if_moves_in_paris(route_detail_dict):
    for segment in route_detail_dict["train_route_details"]:
        start_point = segment["from"]
        end_point = segment["to"]
        if (
            "(Paris)" in start_point
            and "(Paris)" in end_point
            and segment["type"] != "transfer"
        ):
            return True
    return False
```

```python
train_trajets_df["move_in_paris"] = train_trajets_df["route_details_dict"].apply(
    check_if_moves_in_paris
)
```
